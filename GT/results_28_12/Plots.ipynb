# figure 1a: comparison for different values of rho_G_0

## figure 1a for rho_G_0 = 0.0 (closest to paper's plot)

```python
data_1a_5 = pd.read_csv('1a_rhoG0/1a_sigmaG5_rhoG0.txt').as_matrix()
data_1a_6 = pd.read_csv('1a_rhoG0/1a_sigmaG6_rhoG0.txt').as_matrix()
data_1a_7 = pd.read_csv('1a_rhoG0/1a_sigmaG7_rhoG0.txt').as_matrix()
data_1a_8 = pd.read_csv('1a_rhoG0/1a_sigmaG8_rhoG0.txt').as_matrix()
data_1a_9 = pd.read_csv('1a_rhoG0/1a_sigmaG9_rhoG0.txt').as_matrix()
```

```python
#%matplotlib inline
pl.figure(figsize=(10, 6))
plotrange = 40
pl.axis([0, plotrange, -.01, 1])
nFunda_fair_index = 2;
pl.yticks(ticks);
pl.plot([data_1a_5[i][nFunda_fair_index] for i in range(0,plotrange)], label =r'$\sigma_{G_0} = 0.5$', marker = '.')
pl.plot([data_1a_6[i][nFunda_fair_index] for i in range(0,plotrange)], label =r'$\sigma_{G_0} = 0.6$', marker = '.')
pl.plot([data_1a_7[i][nFunda_fair_index] for i in range(0,plotrange)], label =r'$\sigma_{G_0} = 0.7$', marker = '.')
pl.plot([data_1a_8[i][nFunda_fair_index] for i in range(0,plotrange)], label =r'$\sigma_{G_0} = 0.8$', marker = '.')
pl.plot([data_1a_9[i][nFunda_fair_index] for i in range(0,plotrange)], label =r'$\sigma_{G_0} = 0.9$', marker = '.')
pl.legend(fontsize = 'large', loc='center right', bbox_to_anchor=(1.1, .6))
pl.title('Evolution of ' + r'$\rho_F$' ' for ' + r'$\rho_{G_0} = 0.0$', fontsize = 'large')
pl.ylabel(r'$\rho_F$')
pl.xlabel('generation')
#pl.show()
pl.savefig('1a.pdf', bbox_inches='tight')
```
