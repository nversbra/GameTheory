```python
data = pd.read_csv('1a_sigmaG5_rhoG1.txt').as_matrix()
#print(data)
```

```python

%matplotlib inline
pl.figure()
#pl.title('episode reward')
#pl.xlabel('episode')
#pl.ylabel('reward of episode')
plotrange = 40
pl.axis([-0, plotrange, -0.5, 1.5])
pl.plot([data[i][1] for i in range(0,plotrange)], label ='nFunda, modest')
pl.plot([data[i][2] for i in range(0,plotrange)], label ='nFunda, fair')
pl.plot([data[i][3] for i in range(0,plotrange)], label ='nFunda, greedy')
pl.plot([data[i][4] for i in range(0,plotrange)], label ='Funda, modest')
pl.plot([data[i][5] for i in range(0,plotrange)], label ='Funda, fair')
pl.plot([data[i][6] for i in range(0,plotrange)], label ='Funda, greedy')
    
#pl.plot([0,repeats],[qstar,qstar], c='black', label = 'Q*ai',linestyle='dashed')
pl.legend().draggable()
pl.legend(fontsize = 'small', loc='center right', bbox_to_anchor=(1.1, .6))
pl.show()
```
